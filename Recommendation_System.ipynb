```python
# Import Libraries
import pandas as pd

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Load Dataset
df = pd.read_csv("Dataset .csv", encoding='latin-1')
```

```python
# Select Important Columns
df = df[['Restaurant Name', 'Cuisines']]
```

```python
# Remove Missing Values
df.dropna(inplace=True)
```

```python
# Convert Cuisine Text into Matrix
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()

matrix = cv.fit_transform(df['Cuisines'])
```

```python
# Calculate Similarity
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(matrix)
```

```python
# Recommendation Function
def recommend(restaurant):

    index = df[df['Restaurant Name'] == restaurant].index[0]

    distances = similarity[index]

    restaurant_list = sorted(list(enumerate(distances)),
                             reverse=True,
                             key=lambda x:x[1])[1:6]

    for i in restaurant_list:
        print(df.iloc[i[0]]['Restaurant Name'])
```

```python
# Example Recommendation
recommend("Cafe Coffee Day")
```
